## Imports

```python
import pandas as pd

df_election = pd.read_csv('src/election.csv', sep=';')
df_insee = pd.read_csv('src/insee.csv', sep=';', dtype={'Code_INSEE': str})

df_insee_2022 = df_insee[df_insee['Année'] == 2022].copy()
```

```python
df_election.columns
```

## Aperçu des deux dataframes

```python
display(df_election)
display(df_insee_2022)
```

## Suppresion Des Valeurs NaN

```python
null = df_insee.isna().any(axis=1)
display(df_insee[null])
df_insee = df_insee[~null]
```

```python
null = df_election.isna().any(axis=1)
display(df_election[null])
df_election = df_election[~null]
```

## Fusion des deux dataframes

```python
print('Nombre de communes Data Gouv :', len(df_election))
print('Nombre de communes INSEE :', len(df_insee_2022))

df_stats = pd.merge(df_election, df_insee_2022, on=['Année', 'Code_INSEE'], how='inner')
df_model = pd.merge(df_election[['Année', 'Code_INSEE', 'Résultat']], df_insee_2022, on=['Année', 'Code_INSEE'], how='inner')
display(df_model)
```

## Identification des valeurs manquantes

Récupérer un dictionnaire code-libellé depuis les métadonnées du csv population
Et comparer les libellés des communes manquantes avec les libellés présents dans le dict

```python
# Communes présentes dans df_insee mais absentes de df_elections
df_null_election = df_stats[df_stats['Libellé de la commune'].isnull()]
display(df_null_election)

# Communes présentes dans df_election mais absentes de df_insee
df_null_insee = df_stats[df_stats['Population_active'].isnull()]
display(df_null_insee)
```

## Création des CSV

```python
df_stats.to_csv('df_stats.csv', index=False, sep=';', encoding='utf-8')
df_model.to_csv('df_complet.csv', index=False, sep=';', encoding='utf-8')
```
