## Imports

```python
import pandas as pd

df_election = pd.read_csv('src/election.csv', sep=';')
df_insee = pd.read_csv('src/insee.csv', sep=';', dtype={'Code_INSEE': str})

df_insee_2022 = df_insee[df_insee['Année'] == 2022].copy()
```

## Aperçu des deux dataframes

```python
display(df_election)
display(df_insee_2022)
```

## Fusion des deux dataframes

```python
print('Nombre de communes Data Gouv :', len(df_election))
print('Nombre de communes INSEE :', len(df_insee_2022))

df_merged = pd.merge(df_election, df_insee_2022, on=['Année', 'Code_INSEE'], how='outer')
display(df_merged)
```

## Identification des valeurs manquantes

Récupérer un dictionnaire code-libellé depuis les métadonnées du csv population
Et comparer les libellés des communes manquantes avec les libellés présents dans le dict

```python
# Communes présentes dans df_insee mais absentes de df_elections
df_null_election = df_merged[df_merged['Libellé de la commune'].isnull()]
display(df_null_election)


# Communes présentes dans df_election mais absentes de df_insee
df_null_insee = df_merged[df_merged['Population_active'].isnull()]
display(df_null_insee)
```
